```python
import os
import cv2
import pytesseract
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

# Download NLTK data (only need to run once)
nltk.download('punkt')
nltk.download('stopwords')

def extract_text(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform OCR
    text = pytesseract.image_to_string(gray)

    return text

def summarize_text(text, num_sentences=2):
    sentences = sent_tokenize(text)
    stop_words = set(stopwords.words('english'))

    # Create a frequency table of words
    freq_table = {}
    words = word_tokenize(text.lower())
    for word in words:
        if word not in stop_words and word.isalnum():
            freq_table[word] = freq_table.get(word, 0) + 1

    # Score sentences based on word frequencies
    sentence_scores = {}
    for sentence in sentences:
        word_count_in_sentence = len(word_tokenize(sentence))
        for word in word_tokenize(sentence.lower()):
            if word in freq_table:
                if sentence in sentence_scores:
                    sentence_scores[sentence] += freq_table[word]
                else:
                    sentence_scores[sentence] = freq_table[word]
        # Normalize the score by sentence length
        if sentence in sentence_scores:
            sentence_scores[sentence] /= word_count_in_sentence

    # Sort sentences by score
    sorted_sentences = sorted(sentence_scores.items(), key=lambda kv: kv[1], reverse=True)

    # Select top sentences for the summary
    summary_sentences = [sentence for sentence, score in sorted_sentences[:num_sentences]]
    summary = ' '.join(summary_sentences)
    return summary

def process_images_in_folder(folder_path, num_sentences=2, output_folder=None):
    # List all files in the folder
    files = os.listdir(folder_path)

    # Filter out non-image files (you can add more extensions if needed)
    image_extensions = ('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')
    image_files = [f for f in files if f.lower().endswith(image_extensions)]

    # Create output folder if it doesn't exist
    if output_folder and not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        print(f"Processing image: {image_file}")

        # Extract text from the image
        extracted_text = extract_text(image_path)

        # Summarize the extracted text
        summary = summarize_text(extracted_text, num_sentences=num_sentences)

        # Print or save the results
        print("Extracted Text:")
        print(extracted_text)
        print("\nSummary:")
        print(summary)
        print("\n" + "-"*50 + "\n")

        # Optionally, save the extracted text and summary to files
        if output_folder:
            base_filename = os.path.splitext(image_file)[0]
            text_output_path = os.path.join(output_folder, f"{base_filename}_text.txt")
            summary_output_path = os.path.join(output_folder, f"{base_filename}_summary.txt")

            with open(text_output_path, 'w', encoding='utf-8') as f_text:
                f_text.write(extracted_text)

            with open(summary_output_path, 'w', encoding='utf-8') as f_summary:
                f_summary.write(summary)

if __name__ == "__main__":
    # Specify the folder containing the images
    folder_path = 'C:/Users/Zaineb/Downloads/extracted_images1111111'

    # Optionally, specify the output folder
    output_folder = 'C:/Users/Zaineb/Downloads/resume_images'  # Set to None if you don't want to save outputs

    # Number of sentences you want in the summary
    num_summary_sentences = 4

    process_images_in_folder(folder_path, num_sentences=num_summary_sentences, output_folder=output_folder)
```

```python
import os
from PIL import Image, UnidentifiedImageError
import torch
from transformers import DonutProcessor, VisionEncoderDecoderModel


def generate_image_summary(image_path, processor, model):
    try:
        # Load and preprocess the image
        image = Image.open(image_path).convert('RGB')

        # Prepare inputs for the model
        pixel_values = processor(image, return_tensors="pt").pixel_values

        # Generate summary using a task prompt
        task_prompt = "<s_docvqa><s_question>Summarize the content of the image.</s_question><s_answer>"
        input_ids = processor.tokenizer(task_prompt, add_special_tokens=False, return_tensors="pt").input_ids

        # Move tensors to GPU if available
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        model.to(device)
        pixel_values = pixel_values.to(device)
        input_ids = input_ids.to(device)

        # Generate output
        with torch.no_grad():
            outputs = model.generate(
                pixel_values=pixel_values,
                input_ids=input_ids,
                max_length=512,
                early_stopping=True,
                pad_token_id=processor.tokenizer.pad_token_id,
                eos_token_id=processor.tokenizer.eos_token_id,
            )

        # Decode the generated output
        sequence = processor.tokenizer.batch_decode(outputs, skip_special_tokens=False)[0]
        sequence = sequence.replace(processor.tokenizer.eos_token, "").replace(processor.tokenizer.pad_token, "")
        summary = sequence.split("<s_answer>")[-1].strip()
        summary = summary.replace("</s>", "").strip()
        return summary
    except UnidentifiedImageError:
        print(f"Error: Unable to load image at {image_path}. Skipping...")
        return None
    except Exception as e:
        print(f"Error while processing {image_path}: {str(e)}")
        return None


def process_images_in_folder(folder_path, output_folder=None):
    # Load the Donut processor and model once
    processor = DonutProcessor.from_pretrained("naver-clova-ix/donut-base")
    model = VisionEncoderDecoderModel.from_pretrained("naver-clova-ix/donut-base")
    model.eval()

    # List all files in the folder
    files = os.listdir(folder_path)
    image_extensions = ('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')
    image_files = [f for f in files if f.lower().endswith(image_extensions)]

    # Create output folder if it doesn't exist
    if output_folder and not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        print(f"Processing image: {image_file}")

        # Generate summary
        summary = generate_image_summary(image_path, processor, model)

        # If summary generation was successful
        if summary:
            # Print or save the results
            print(f"Summary for {image_file}:")
            print(summary)
            print("\n" + "-"*50 + "\n")

            # Optionally, save the summary to a file
            if output_folder:
                base_filename = os.path.splitext(image_file)[0]
                summary_output_path = os.path.join(output_folder, f"{base_filename}_summary.txt")

                with open(summary_output_path, 'w', encoding='utf-8') as f_summary:
                    f_summary.write(summary)
        else:
            print(f"Skipping {image_file} due to an error.\n" + "-"*50 + "\n")


if __name__ == "__main__":
    folder_path = 'C:/Users/Zaineb/Downloads/extracted_images1111111'  # Update with your folder path
    output_folder = 'C:/Users/Zaineb/Downloads/resume_images4'        # Update with your output folder

    process_images_in_folder(folder_path, output_folder=output_folder)
```

```python
import os
import pytesseract
import cv2
from transformers import pipeline
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu

def generate_image_summary(image_path, summarizer):
    try:
        # Load and preprocess the image
        image = cv2.imread(image_path)
        if image is None:
            print(f"Error: Unable to read image at {image_path}. Skipping...")
            return None
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Extract text using OCR
        text = pytesseract.image_to_string(image)

        if not text.strip():
            print("No text found in image.")
            return None

        # Summarize the extracted text
        summary = summarizer(text, max_length=150, min_length=40, do_sample=False)[0]['summary_text']
        return summary
    except Exception as e:
        print(f"Error processing {image_path}: {str(e)}")
        return None

# Function to calculate ROUGE and BLEU scores for evaluation
def evaluate_summary(generated_summary, reference_summary):
    # Initialize ROUGE scorer
    rouge = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    rouge_scores = rouge.score(reference_summary, generated_summary)

    # Calculate BLEU score
    reference_tokens = [reference_summary.split()]
    generated_tokens = generated_summary.split()
    bleu_score = sentence_bleu(reference_tokens, generated_tokens)

    return rouge_scores, bleu_score

def process_images_in_folder(folder_path, reference_summaries, output_folder=None):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    files = os.listdir(folder_path)
    image_extensions = ('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')
    image_files = [f for f in files if f.lower().endswith(image_extensions)]

    if output_folder and not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for idx, image_file in enumerate(image_files):
        image_path = os.path.join(folder_path, image_file)
        print(f"Processing image: {image_file}")

        summary = generate_image_summary(image_path, summarizer)

        if summary:
            print(f"Summary for {image_file}:")
            print(summary)

            # Get the reference summary for this image
            reference_summary = reference_summaries[idx]

            # Evaluate the generated summary
            rouge_scores, bleu_score = evaluate_summary(summary, reference_summary)
            print(f"ROUGE Scores for {image_file}: {rouge_scores}")
            print(f"BLEU Score for {image_file}: {bleu_score}")
            print("\n" + "-"*50 + "\n")

            if output_folder:
                base_filename = os.path.splitext(image_file)[0]
                summary_output_path = os.path.join(output_folder, f"{base_filename}_summary.txt")

                with open(summary_output_path, 'w', encoding='utf-8') as f_summary:
                    f_summary.write(summary)
        else:
            print(f"Skipping {image_file} due to an error.\n" + "-"*50 + "\n")

if __name__ == "__main__":
    folder_path = 'C:/Users/Zaineb/Downloads/extracted_images1111111'
    output_folder = 'C:/Users/Zaineb/Downloads/resume_images5'
    
    # Manually prepared reference summaries for evaluation (corresponding to each image)
    reference_summaries = [
        "This is the first reference summary.",
        "This is the second reference summary.",
        # Add more summaries for each image
    ]

    process_images_in_folder(folder_path, reference_summaries, output_folder=output_folder)
```
